```python
def get_coding_sequences(ORF_file):
    f = open(ORF_file,'r')
    out_name = 'coding_' + ORF_file[ORF_file.find('/',2)+1:]
    f_out = open(out_name, 'w')
    head = f.readline()
    f_out.write(head)
    for line in f:
        if line.split(',')[5] == 'coding' and float(line.split(',')[6]) >= 0.9:
            f_out.write(line)
    print(ORF_file + ' done')
        
    
```
