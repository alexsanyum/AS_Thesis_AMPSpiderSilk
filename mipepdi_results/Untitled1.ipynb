```python
from Bio.Seq import Seq

def get_aa_seqs(ORF_mipepid):
    f = open(ORF_mipepid)
    f.readline()
    aa_list = []
    for line in f:
        data = line.split(',')
        entry = [data[0], data[1], str(Seq(data[1]).translate())]
        aa_list.append(entry)
    return aa_list
```

```python
aa_L_hesperus = get_aa_seqs('coding_ORF_Latrodectushesperus_no_clstrd_no_N.csv')
```

```python
import os
files = os.listdir()
all_list = []
for file in files:
    if 'coding' in file:
        all_list.append(get_aa_seqs(file))
        
        
```
