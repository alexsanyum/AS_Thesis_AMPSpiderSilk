```python
from Bio.Seq import Seq

def get_aa_seqs(ORF_mipepid):
    f = open(ORF_mipepid)
    f.readline()
    aa_list = []
    for line in f:
        data = line.split(',')
        entry = [data[0], data[1], str(Seq(data[1]).translate())]
        aa_list.append(entry)
    return aa_list
```

```python
import os 
all_list = []
for file in os.listdir('./no_clustered/'):
    if 'coding' in file and 'trimed' in file:
        all_list.append([file, get_aa_seqs('./no_clustered/'+file)])
```

```python

all_list[0]
```

```python
for specie in all_list:
    out_name = specie[0][:specie[0].find('.csv')] + '_prot.fasta'
    f_out = open(out_name, 'w')
    for line in specie[1]:
        if len(line[2].strip('*')) > 5:
            f_out.write('>'+line[0]+'\n')
            f_out.write(line[2].strip('*')+ '\n')
    f_out.close()
        
```
