```python
def combine_all(files_names):
    f_out = open('all_seqs.fasta', 'w')
    for file in files_names:
        f = open(file,'r')
        for line in f:
            f_out.write(line)
        f.close
    print('All sequences were merged in all_seqs.fasta')
```

```python
import os
files = os.listdir('no_N_sequences/')
c_files = os.listdir('with_N_seqs/')
path_files_no_clstrd = ['no_N_sequences/' + filename for filename in files if '.fasta' in filename]
path_files_consensus = ['with_N_seqs/' + filename for filename in c_files if '.fasta' in filename]
all_files = path_files_consensus + path_files_no_clstrd
```

```python
combine_all(all_files)
```

```python
def count_seq_names(fasta_file):
    f = open(fasta_file, 'r')
    seq_names = []
    for line in f:
        if '>' not in line:
            seq_names.append(line)
    seq_no_duplicate = list(set(seq_names))
    print(len(seq_names))
    print(len(seq_no_duplicate))
```

```python
count_seq_names('all_seqs.fasta')
```
