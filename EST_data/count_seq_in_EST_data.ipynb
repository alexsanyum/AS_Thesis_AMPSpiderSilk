```python
def count_seq_names(fasta_file):
    f = open(fasta_file, 'r')
    seq_names = []
    for line in f:
        if '>' not in line:
            seq_names.append(line.rstrip())
    seq_no_duplicate = list(set(seq_names))
    print(len(seq_names))
    print(len(seq_no_duplicate))
```

```python
count_seq_names('Latrodectushesperus.fasta')
```
