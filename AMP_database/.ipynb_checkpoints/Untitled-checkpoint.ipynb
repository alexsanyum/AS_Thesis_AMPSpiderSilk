```python
def separe_seq_by_length(fasta_file, min_len, max_len, increase):
#     f = open(fasta_file)
    len_range = range(min_len, max_len+1, increase)
    for i in range(len(len_range)-1):
        out_name = str(len_range[i]) + '-' + str(len_range[i+1])
        print(out_name)
```
