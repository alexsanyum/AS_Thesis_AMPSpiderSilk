```python
import re
def delete_white_lines(file):
    f = open(file)
    lines = f.readlines()
    f.close()
    lines_no_white = []
    for line in lines:
        if line != '\n':
            lines_no_white.append(line)
    return lines_no_white
def get_dict_blast(blast_output):
    blst_lines = delete_white_lines(blast_output)
    blst_dict = {}
    for line in blst_lines:
        if 'Query=' in line:
            blst_key = line[line.find('Query')+7:line.find('\n')]
            blst_dict[blst_key] = []
        elif 'No hits found' in line:
            blst_dict[blst_key] = ['No hits found']
        elif line[0] == 'L' and '|' in line:
            entry = line.rstrip()
            first_split = re.split('\|',entry)
            hit_id = first_split[0]
            hit_name, hit_score, hit_e_value = re.split('\s\s+', re.split('\|',entry)[1])
            blst_value = [hit_id,hit_name, hit_score, hit_e_value]
            blst_dict[blst_key].append(blst_value)
    return blst_dict
def get_info_blast(blast_output):
    dict_blast = get_dict_blast(blast_output)
    no_hit_count = 0
    for key in dict_blast.keys():
        if dict_blast[key][0] == 'No hits found':
            no_hit_count += 1
        else:
            print(key)
            for line in dict_blast[key]:
                print(line)
    print('No hit sequences: ' + str(no_hit_count))
```

```python
import os
no_clustered_list = os.listdir('./no_clustered/')

for file in no_clustered_list:
    path_file = './no_clustered/' + file 
    get_info_blast(path_file)
```

```python
consensus_list = os.listdir('./consensus/')
for file in consensus_list:
    path_file = './consensus/' + file 
    get_info_blast(path_file)
```
