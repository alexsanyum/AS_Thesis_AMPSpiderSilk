```python
import os

def join_seq_sequences(list_fasta_seqs):
    dict_files = det_fasta_names(list_fasta_seqs)
    for key in list(dict_files.keys()):
        f_out = open(key,'w')
        for fasta_file in dict_files[key]:
            f_in = open(fasta_file,'r')
            f_in_lines = f_in.readlines()
            fasta_id = '>' + fasta_file[:fasta_file.find('.fasta')]
            seq = f_in_lines[1].rstrip()
            f_out.write(fasta_id + '\n')
            f_out.write(seq + '\n')
            f_in.close()
        f_out.close()
        

def det_fasta_names(files_names):
    list_names = {}
    for file in files_names:
        if '.fa' in file:
            file_name = file[:file.find('0_75')-1] + '.fasta'
            if file_name not in list_names.keys():
                list_names[file_name] = [file]
            else:
                list_names[file_name].append(file)
    return list_names
    
```
