```python
def get_reverse_seq(fasta_file):
    base_dict = {'A':'T', 'C':'G', 'G':'C', 'T':'A'}
    f = open(fasta_file,'r')
    out_name = fasta_file[:fasta_file.find('.fasta')] + '_reverse.fasta'
    f_out = open(out_name, 'w')
    complement = ''
    for line in f:
        if line[0] != '>':
            for base in line.rstrip():
                complement = base_dict[base] + complement
            f_out.write(complement + '\n')
        else:
            f_out.write(line)
    return fasta_file + ' done'
            
```

```python
import os
list_files = os.listdir()
```
