```python
def retire_with_N_sequences(fasta_file):
    f = open(fasta_file,'r')
    out_name_no_N = fasta_file[:fasta_file.find('.fas')] + 'no_N.fasta'
    out_name_with_N = fasta_file[:fasta_file.find('.fas')] + 'with_N.fasta'
    f_no_N = open(out_name_no_N,'w')
    f_with_N = open(out_name_with_N,'w')
    for line in f:
        if line[0] == '>':
            seqID = line
        elif 'N' in line.rstrip():
            f_with_N.write(seqID)
            f_with_N.write(line)
        else:
            f_no_N.write(seqID)
            f_no_N.write(line)
        print(fasta_file, ' done')
            
def separe_all(files_list):
    for file in files_list:
        retire_with_N_sequences(file)
```

```python
import os
files = os.listdir()
del(files[0])
del(files[-1])
separe_all(files)
```
